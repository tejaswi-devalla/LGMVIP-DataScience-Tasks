```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
tracks = pd.read_csv("members.csv")
tracks
```

```python
songs = pd.read_csv("songs.csv")
songs
```

```python
train = pd.read_csv("train.csv")
train
```

```python
test = pd.read_csv('test.csv')
test
```

```python
new_data = train.merge(songs[['song_id','song_length','genre_ids','artist_name','language']], on=['song_id'], how='left')
new_data
```

```python
songs_extra = pd.read_csv("song_extra_info.csv")
songs_extra
```

```python
new_songs = new_data.merge(songs_extra,on=['song_id'],how = 'left')
new_songs
```

```python
song_id = new_songs.loc[:,["name","target"]]
song = song_id.groupby(["name"],as_index=False).count().rename(columns = {"target":"tot_listen"})
song
```

```python
dataset=new_songs.merge(song,on=['name'],how= 'left')
dataset
```

```python
df=pd.DataFrame(dataset)
```

```python
df.drop(columns=['source_system_tab','source_screen_name','source_type','target','isrc'],axis=1,inplace=True)
df=df.rename(columns={'msno':'user_id'})
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['song_length'].fillna('0',inplace=True)
df['genre_ids'].fillna('0',inplace=True)
df['artist_name'].fillna('none',inplace=True)
df['language'].fillna('0',inplace=True)
df['name'].fillna('none',inplace=True)
df['tot_listen'].fillna('0',inplace=True)
```

```python
df.isnull().sum()
```

```python
df['song'] = df['name'].map(str) + " - " + df['artist_name']
```

```python
song_gr = df.groupby(['song']).agg({'tot_listen': 'count'}).reset_index()
grouped_sum = song_gr['tot_listen'].sum()
song_gr['percentage']  = song_gr['tot_listen'].div(grouped_sum)*100
song_gr.sort_values(['tot_listen', 'song'], ascending = [0,1])
```
